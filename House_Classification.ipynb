```python
import pandas as pd
from sklearn.model_selection import train_test_split # pembagi dataset menjadi training dan testing set
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report # evaluasi performa model
```

```python
house = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/house_data.csv')
house.head()
```

```python
house.drop(columns='date', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='bathrooms', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='sqft_living', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='sqft_lot', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='floors', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='waterfront', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='view', inplace=True) # menghapus kolom bernama 'Id'
house.head() # tampilkan 5 baris pertama
```

```python
house.drop(columns='condition', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='grade', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='sqft_above', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='sqft_basement', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='yr_built', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='yr_renovated', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='zipcode', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='lat', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='long', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='sqft_living15', inplace=True) # menghapus kolom bernama 'Id'
house.drop(columns='sqft_lot15', inplace=True) # menghapus kolom bernama 'Id'


house.head() # tampilkan 5 baris pertama
```

```python
X = house.drop(columns='id') # menempatkan features ke dalam variable X
X.head() # tampilkan 5 baris pertama
```

```python
y = house['id'] # menempatkan class label (target) ke dalam variabel y
y.head() # tampilkan 5 baris pertama
```

```python
# membagi dataset ke dalam training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)

print('training dataset')
print(X_train.shape)
print(y_train.shape)
print()
print('testing dataset:')
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_range = list(range(1,26))
scores = []
for k in k_range:
    model_knn = KNeighborsClassifier(n_neighbors=k) # konfigurasi algoritma
    model_knn.fit(X_train, y_train) # training model/classifier
    y_pred = model_knn.predict(X_test) # melakukan prediksi
    scores.append(accuracy_score(y_test, y_pred)) # evaluasi performa
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.tight_layout()
plt.show()
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3) # konfigurasi algoritma
model_knn.fit(X_train,y_train) # training model/classifier
y_pred = model_knn.predict(X_test) # melakukan prediksi
```

```python
print(accuracy_score(y_test, y_pred)) # evaluasi akurasi
```

```python
print(confusion_matrix(y_test, y_pred)) # evaluasi confusion matrix
```

```python
print(classification_report(y_test, y_pred)) # evaluasi klasifikasi
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# model_logreg = LogisticRegression()
model_logreg = LogisticRegression(solver='lbfgs', multi_class='auto')
model_logreg.fit(X_train,y_train)
y_pred = model_logreg.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
```

```python
# model_svc = SVC()
model_svc = SVC(gamma='scale')
model_svc.fit(X_train,y_train)
y_pred = model_svc.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
y_pred = model_dt.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# model_rf = RandomForestClassifier()
model_rf = RandomForestClassifier(n_estimators=100)
model_rf.fit(X_train,y_train)
pred_rf = model_rf.predict(X_test)
```

```python
models = [model_knn, model_logreg, model_svc, model_dt, model_rf]
accuracy_scores = []
for model in models:
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    
```

```python
plt.bar(['KNN', 'LogReg', 'SVC', 'DT', 'RF'],accuracy_scores)
plt.ylim(0.10,1.01)
plt.title('Accuracy comparision for various models', fontsize=15, color='r')
plt.xlabel('Models', fontsize=18, color='g')
plt.ylabel('Accuracy Score', fontsize=18, color='g')
plt.tight_layout()
plt.show()
```
